```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## For line 0

### Read 31st August's data from csv file

```python
data0 = pd.read_csv("31Aug.csv")
```

### Extract data values from 31st August

```python
#g observed
gobs0 = data0["G(obs)"].values

#time in hour
time_hr0 = data0["Time(hr)"].values

#time in min
time_min0 = data0["Time(min)"].values

#station number
stn0  = data0["Station"].values
```

```python
#number of rows
lr0 = data0.shape[0]

T0 = []

#Iterate through all the data points
for i in range(0,lr0):
    s = (time_min0[i] - time_min0[0]) + (60.*(time_hr0[i] - time_hr0[0]))
    #Append the time(in minutes) to Array
    T0.append(s)
```

### Correct g(observed) and make an array of drifts

```python
#Gradient
grad0 = (gobs0[lr0-1] - gobs0[0]) / (T0[lr0-1] - T0[0])

gcorr0 = []
drift0 = []
for j in  range(0,lr0):
    gcorr0.append(gobs0[j] - grad0*T0[j])
    drift0.append(gobs0[j] -  gcorr0[j])
```

### Plot Line 0

```python
fig, ax = plt.subplots(figsize = (22,10))
ax.plot(T0, gobs0, '*-', label='Observed value')
ax.plot(T0, gcorr0, '.-', label='Drift Corrected value')
ax.legend(fontsize = 25)
ax.annotate((stn0[0],gcorr0[0]) , xy = (T0[0],gcorr0[0])).set_fontsize(20)
ax.annotate((stn0[-1],gcorr0[-1]) , xy = (T0[-1],gcorr0[-1])).set_fontsize(20)
ax.annotate((stn0[14],gcorr0[14]) , xy = (T0[14],gcorr0[14])).set_fontsize(20)
ax.annotate((stn0[29],gcorr0[29]) , xy = (T0[29],gcorr0[29])).set_fontsize(20)
plt.xlabel("Time Passed since first reading (in minutes)").set_fontsize(25)
plt.ylabel("g value (mGal)").set_fontsize(25)
plt.title("Observed and Drift Corrected Gravity Values for Line 0").set_fontsize(25)
plt.savefig('Results/L0-1.png')
```

```python
plt.figure(figsize=(15,8))
plt.plot(T0, drift0, '.-')
plt.xlabel("Time passed since first reading (in minutes)").set_fontsize(20)
plt.ylabel("Drift value (mGal)").set_fontsize(20)
plt.title("Drift Curve for Line 0 on August 31").set_fontsize(20)
plt.savefig('Results/L0-2.png')
```
